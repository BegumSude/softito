# 🤖 KNN vs Naive Bayes — Kapsamlı Karşılaştırma
### Iris Çiçek Türleri ile Sınıflandırma Analizi

---

**Veri Seti:** Iris Dataset — Çiçeklerin çanak ve taç yaprak boyutlarına göre türlerinin (Setosa, Versicolor, Virginica) tahmin edilmesi

## 📚 1. KNN (K-Nearest Neighbors) Nedir?

**KNN**, bir örneği sınıflandırmak için eğitim verisindeki **en yakın K komşuya** bakan basit ama güçlü bir algoritmadır.

- **Lazy Learning**: Eğitim sırasında hiçbir şey öğrenmez, tüm işlem tahmin aşamasında yapılır.
- **Instance-based**: Her tahmin için tüm eğitim verisi kullanılır.
- **Non-parametric**: Verinin dağılımı hakkında hiçbir varsayımda bulunmaz.

### Nasıl Çalışır?
1. Yeni örnek gelir
2. Tüm eğitim örneklerine uzaklık hesapla
3. En yakın K örneği seç
4. Bu K komşunun çoğunluk sınıfını ata

## 📊 2. Naive Bayes Nedir?

**Naive Bayes**, Bayes Teoremi'ne dayanan ve tüm özelliklerin birbirinden **bağımsız** olduğunu varsayan (bu yüzden "naive" yani saf denilen) olasılıksal bir sınıflandırma algoritmasıdır.

### Bayes Teoremi:
$$P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)}$$

### 🔎 Türleri:
- **Gaussian NB**: Özellikler sürekli/sayısal veri olduğunda kullanılır (Biz bu çalışmada bunu kullanacağız).
- **Multinomial NB**: Metin sınıflandırmasında (kelime sayımları) kullanılır.
- **Bernoulli NB**: İkilik (binary) özelliklerde (var/yok) kullanılır.

---
## 3. Gerekli Kütüphanelerin Yüklenmesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report

print("Kutuphaneler basariyla yuklendi.")
```

---
## 4. Veri Setinin Yüklenmesi ve Keşfi

Bu çalışmada, makine öğrenmesi dünyasının en klasik ve popüler çoklu sınıflandırma veri seti olan **Iris (Çiçek Verisi)** kullanılacaktır.

- **Özellikler (Features):** Çanak yaprak uzunluğu (`sepal length`), çanak yaprak genişliği (`sepal width`), taç yaprak uzunluğu (`petal length`), taç yaprak genişliği (`petal width`).
- **Hedef Değişken (Target):** Çiçek türü (0: Setosa, 1: Versicolor, 2: Virginica).

```python
# Scikit-learn içerisindeki hazır temiz Iris veri setini yüklüyoruz
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

print("Veri Seti İlk 5 Satır:")
display(df.head())

print(f"\nVeri Seti Boyutu: {df.shape}")
print("\nSınıf Dağılımları:")
print(df['target'].value_counts())
```

---
## 5. Verinin Eğitim ve Test Olarak Ayrılması

Veriyi orijinal şablona sadık kalarak %80 Eğitim, %20 Test olacak şekilde ayırıyoruz.

```python
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print(f"Eğitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")
```

---
## 6. Verilerin Ölçeklendirilmesi (Feature Scaling)

KNN algoritması Öklid mesafe hesabına dayandığı için ölçeklendirme adımı uygulanmaktadır.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Ölçeklendirme işlemi başarıyla tamamlandı.")
```

---
## 7. KNN Modelinin Eğitilmesi ve Performansı

```python
# Komşu sayısı k=5 seçilmiştir
knn_model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_model.fit(X_train_scaled, y_train)

y_pred_knn = knn_model.predict(X_test_scaled)

print("=== KNN Sınıflandırma Raporu ===")
print(classification_report(y_test, y_pred_knn))
```

---
## 8. Naive Bayes Modelinin Eğitilmesi ve Performansı

```python
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)

y_pred_nb = nb_model.predict(X_test_scaled)

print("=== Naive Bayes Sınıflandırma Raporu ===")
print(classification_report(y_test, y_pred_nb))
```

---
## 9. Karmaşıklık Matrislerinin (Confusion Matrix) Karşılaştırılması

```python
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_nb = confusion_matrix(y_test, y_pred_nb)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# KNN Heatmap
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=iris.target_names, yticklabels=iris.target_names)
axes[0].set_title('KNN Confusion Matrix')
axes[0].set_xlabel('Tahmin Edilen Sınıf')
axes[0].set_ylabel('Gerçek Sınıf')

# Naive Bayes Heatmap
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Greens', ax=axes[1],
            xticklabels=iris.target_names, yticklabels=iris.target_names)
axes[1].set_title('Naive Bayes Confusion Matrix')
axes[1].set_xlabel('Tahmin Edilen Sınıf')
axes[1].set_ylabel('Gerçek Sınıf')

plt.tight_layout()
plt.show()
```

---
## 10. Modellerin Metrik Karşılaştırma Raporu

```python
metrics_data = {
    'Metrik': ['Doğruluk (Accuracy)', 'Hassasiyet (Precision)', 'Duyarlılık (Recall)', 'F1-Skoru'],
    'KNN Model': [
        accuracy_score(y_test, y_pred_knn),
        precision_score(y_test, y_pred_knn, average='macro'),
        recall_score(y_test, y_pred_knn, average='macro'),
        f1_score(y_test, y_pred_knn, average='macro')
    ],
    'Naive Bayes Model': [
        accuracy_score(y_test, y_pred_nb),
        precision_score(y_test, y_pred_nb, average='macro'),
        recall_score(y_test, y_pred_nb, average='macro'),
        f1_score(y_test, y_pred_nb, average='macro')
    ]
}

df_metrics = pd.DataFrame(metrics_data)
print("=== PERFORMANS KARŞILAŞTIRMA TABLOSU ===")
display(df_metrics.round(4))
```

## 🚀 11. Özet ve Seçim Kriterleri

| Özellik | KNN | Naive Bayes |
|---|---|---|
| **En iyi durum** | Küçük, temiz, ölçeklenmiş veri | Büyük veri, metin, gerçek zamanlı |
| **Güçlü yönü** | Varsayımsız, esnek sınır | Hız, basitlik, olasılık çıktısı |
| **Zayıf yönü** | Bellek & tahmin yavaşlığı | Bağımsızlık varsayımı |
| **Ölçekleme** | **Zorunlu** | Gerekli değil |

> 💡 **Kural:** KNN sezgisel ve esnek, Naive Bayes hızlı ve yorumlanabilir.
